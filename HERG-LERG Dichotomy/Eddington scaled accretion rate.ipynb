# Obtaining the ratio of black hole mass to the solar mass using the velocity distribution in the catalog, and henceforth calculating the Eddington Luminosity

## The two above dataframes were merged on the basis of matching RA and DEC coordinates in TOPCAT software and the resultant table is used here (rogue_all.csv)

```python
import numpy as np
import pandas as pd

rogue1 = pd.read_csv("radio_loud4.csv")
rogue = rogue1[rogue1.El_F_5007 != -999]
rogue = rogue.set_index("No.")

vd = rogue["vd"].tolist()
n = len(rogue)

lg_mass_ratio = [] #M_BH/M_Solar
L_edd = []
mass_ratio = []

for i in range(n):
    
    mass_ratio1 = pow(10,(8.13 + 4.02*np.log10(vd[i]/200)))
    mass_ratio.append(mass_ratio1)
    lg_mass_ratio.append((8.13 + 4.02*np.log10(vd[i]/200)))
    L_edd.append(1.3*(10**31)*mass_ratio1)
    
display(lg_mass_ratio)

```

# Calculating the Mechanical Luminosity ($L_{mech}$) 
## $L_{1.4} = \frac{S_{obs}4 \pi {D_L}^2}{(1 + z)^{(1 + \alpha)}}$
## $L_{mech} = 7.3 \times 10^{36} \times \left(L_{1.4} \times 10^{-24}\right) ^ {0.7}$


```python
import pandas as pd
import numpy as np

n = len(rogue)

flux1 = rogue["TotalFlux"].tolist()
DL1 = rogue["DL"].tolist()

DL = [m *3.08* (10**22) for m in DL1]
flux = [k * (10**(-29)) for k in flux1]
    

#print(DL)

z = rogue['z'].tolist()


alpha_r = -0.3

L_mech = []
lg_L14 = []

for i in range(n):    
    L_14 = (flux[i]*4*np.pi*(DL[i]**2))/(pow((1+z[i]),(1 + alpha_r)))
    lg_L14.append(np.log10(L_14))
    L_m = 7.3*(10**36)*((L_14*(10**(-24)))**(0.70))
    L_mech.append(L_m)
    
display(lg_L14)
```

# Calculating the Radiative Luminosity ($L_{rad}$)
## $L_{(OIII)} = S_{(OIII)}4 \pi {D_L}^2$
## $L_{rad} = 3500L_{(OIII)}$

```python
import pandas as pd
import numpy as np

n = len(rogue)

S_O3_1 = rogue["El_F_5007"].tolist()
DL1 = rogue["DL"].tolist()

DL = [m *3.08* (10**22) for m in DL1]
S_OIII = [k * (10**(-24)) for k in S_O3_1]

L_OIII = []

for i in range(n):
    L_1 = (S_OIII[i]*4*np.pi*(DL[i]**2))
    L_OIII.append(L_1)
    
L_rad = []

for j in range(n):
    L_2 = 3500*L_OIII[j]
    L_rad.append(L_2)
    
    
display(L_rad)
```

# Calculating the Eddington Scaled Accretion Rate ($\lambda$) and finding the logarithmic value ($\log_{10}{(\lambda)}$)
## $\lambda = \frac{(L_{mech} + L_{rad})}{L_{Edd}}$

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

acc_rate = []
log_acc_rate = []

m = len(rogue)

for i in range(m):
    rate = ((L_rad[i] + L_mech[i])/L_edd[i])
    log_rate = np.log10(rate)
    acc_rate.append(rate)    
    log_acc_rate.append(log_rate)
    

rogue['acc_rate)'] = acc_rate
rogue['log_acc_rate'] = log_acc_rate

rogue.to_csv('rl_final4.csv')

display(rogue)

```

# Plotting the histogram showing the distribution of the (logarithmic) Eddington Scalled Accretion rate

```python


acc_rate = rogue['log_acc_rate'].tolist()


plt.figure(figsize=(10,6))

plt.hist(acc_rate,bins = 'auto', range = [-5,0], color='blue', histtype = 'step', label = 'ROGUE-I Sources', density = True)

plt.legend()
plt.xlabel('$log_{10}{([L_{//Mech} + L_{Rad}] / {L_{Edd}})}$')
plt.ylabel('Number of sources')
plt.title('Histogram depicting the distribution of the eddington scaled\naccretion rates for the ROGUE-I sources (logarithmic)\n')

plt.show()
```
