# Ensuring that all the rows contain the required line ratios, thus eliminating the NaN values (-999)

```python
import numpy as np
import pandas as pd

rogue1 = pd.read_csv("rl_final4.csv")
rogue = rogue1[(rogue1.O3Hb_dr != -999) & (rogue1.N2Ha_dr != -999) & (rogue1.S2Ha_dr != -999) & (rogue1.O1Ha_dr != -999)]
rogue = rogue.set_index("No.")

display(rogue1)

display(rogue)
```

# For the sources obtained in the previous cell, calculating the Excitation Index


```python
import pandas as pd
import numpy as np
import statistics as stat

n = len(rogue)

o3hb = rogue["O3Hb_dr"].tolist()
n2ha = rogue["N2Ha_dr"].tolist()
s2ha = rogue["S2Ha_dr"].tolist()
o1ha = rogue["O1Ha_dr"].tolist()


EI = []

for i in range(n):
    EI.append(o3hb[i] - (1/3)*(n2ha[i] + s2ha[i] + o1ha[i]))

    
rogue['EI'] = EI

sigma = stat.stdev(EI)
display(rogue)
print("Standard Deviation: ",sigma)
```

```python
herg = rogue[rogue.EI > (0.95)]
display(herg)
herg.to_csv("herg4.csv")

common = rogue1.merge(herg, on=["RAdeg","DEdeg"])
lerg = rogue1[(~rogue1.RAdeg.isin(common.RAdeg)) & (~rogue1.DEdeg.isin(common.DEdeg))]

display(lerg)
lerg.to_csv("lerg4.csv")
```

###$D_n(4000)$ vs $log_{10}{(L_{1.4})}$ 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


herg = pd.read_csv("herg4.csv")
lerg = pd.read_csv("lerg4.csv")


m = len(herg)
n = len(lerg)



fluxh1 = herg["TotalFlux"].tolist()
DLh1 = herg["DL"].tolist()

DLh = [l *3.08* (10**22) for l in DLh1]
fluxh = [k * (10**(-29)) for k in fluxh1]


fluxl1 = lerg["TotalFlux"].tolist()
DLl1 = lerg["DL"].tolist()

DLl = [l *3.08* (10**22) for l in DLl1]
fluxl = [k * (10**(-29)) for k in fluxl1]


zh = herg['z'].tolist()
zl = lerg['z'].tolist()


alpha_r = -0.3

lg_L14h = []
lg_L14l = []

for i in range(m):    
    lg_L14h.append(np.log10(fluxh[i]*4*np.pi*(DLh[i]**2))/((1+zh[i])**(1 + alpha_r)))

for i in range(n):    
    lg_L14l.append(np.log10(fluxl[i]*4*np.pi*(DLl[i]**2))/((1+zl[i])**(1 + alpha_r)))


herg['lg_L14h'] = lg_L14h
lerg['lg_L14l'] = lg_L14l


d4000_h = herg["D4000"].tolist()
d4000_l = lerg["D4000"].tolist()


total_bins = 6

bins1 = np.geomspace(min(lg_L14l), max(lg_L14l), total_bins)
cut1 = pd.cut(lerg.lg_L14l,bins1)         
grp1 = lerg.groupby(by = cut1)        
ret1 = grp1.aggregate(np.mean)         

bins2 = np.geomspace(min(lg_L14h), max(lg_L14h), total_bins)
cut2 = pd.cut(herg.lg_L14h,bins2)         
grp2 = herg.groupby(by = cut2)        
ret2 = grp2.aggregate(np.mean)         



plt.figure(figsize=(10,6))


coeff1 = np.polyfit(ret1.lg_L14l, ret1.D4000, 1)
coeff2 = np.polyfit(ret2.lg_L14h, ret2.D4000, 1)

poly1d_fn1 = np.poly1d(coeff1)
poly1d_fn2 = np.poly1d(coeff2)

plt.plot(lg_L14l, d4000_l, color = 'darkblue', marker = ".", markersize = 1, linestyle='None', alpha = 0.5)
plt.plot(lg_L14h, d4000_h, color = 'red', marker = ".", markersize = 2, linestyle='None')


plt.plot(ret1.lg_L14l,ret1.D4000, color = 'darkblue', marker = "o", markersize = 15, linestyle = 'None', alpha = 0.7)
plt.plot(ret2.lg_L14h,ret2.D4000, color = 'red', marker = "o", markersize = 15, linestyle = 'None', alpha = 0.7)


plt.errorbar(ret1.lg_L14l,ret1.D4000, color = 'blue', marker = "None", linestyle = 'None', yerr=poly1d_fn1(ret1.lg_L14l) - ret1.D4000, linewidth = 3, label = 'Low Excitation')
plt.errorbar(ret2.lg_L14h,ret2.D4000, color = 'red', marker = "None", linestyle = 'None', yerr=poly1d_fn2(ret2.lg_L14h) - ret2.D4000, linewidth = 3, label = 'High Excitation')


plt.xlabel('$log_{10}{(L_{1.4}}$ / $W Hz^{-1})$')
plt.ylabel('$D_n(4000)$')
plt.grid()
plt.legend()
plt.show()
```

##$(R_{90}$ / $R_{50}$) vs $log_{10}{(L_{1.4})}$ 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


herg = pd.read_csv("herg4.csv")
lerg = pd.read_csv("lerg4.csv")


m = len(herg)
n = len(lerg)

display(m)
display(n)

fluxh1 = herg["TotalFlux"].tolist()
DLh1 = herg["DL"].tolist()

DLh = [m *3.08* (10**22) for m in DLh1]
fluxh = [k * (10**(-29)) for k in fluxh1]


fluxl1 = lerg["TotalFlux"].tolist()
DLl1 = lerg["DL"].tolist()

DLl = [m *3.08* (10**22) for m in DLl1]
fluxl = [k * (10**(-29)) for k in fluxl1]


zh = herg['z'].tolist()
zl = lerg['z'].tolist()


alpha_r = -0.3

lg_L14h = []
lg_L14l = []

for i in range(m):    
    lg_L14h.append(np.log10(fluxh[i]*4*np.pi*(DLh[i]**2))/((1+zh[i])**(1 + alpha_r)))

for i in range(n):    
    lg_L14l.append(np.log10(fluxl[i]*4*np.pi*(DLl[i]**2))/((1+zl[i])**(1 + alpha_r)))


herg['lg_L14h'] = lg_L14h
lerg['lg_L14l'] = lg_L14l


r90h = herg['R90'].tolist()
r50h = herg['R50'].tolist()


r90l = lerg['R90'].tolist()
r50l = lerg['R50'].tolist()


C_h = []
C_l = []

for i in range(m):
    C_h.append(r90h[i]/r50h[i])
    
for j in range(n):
    C_l.append(r90l[j]/r50l[j])

herg['C_h'] = C_h
lerg['C_l'] = C_l


total_bins = 6

bins1 = np.geomspace(min(lg_L14l), max(lg_L14l), total_bins)
cut1 = pd.cut(lerg.lg_L14l,bins1)         
grp1 = lerg.groupby(by = cut1)        
ret1 = grp1.aggregate(np.mean)         

bins2 = np.geomspace(min(lg_L14h), max(lg_L14h), total_bins)
cut2 = pd.cut(herg.lg_L14h,bins2)         
grp2 = herg.groupby(by = cut2)        
ret2 = grp2.aggregate(np.mean)         



plt.figure(figsize=(10,6))



coeff1 = np.polyfit(ret1.lg_L14l, ret1.C_l, 1)
coeff2 = np.polyfit(ret2.lg_L14h, ret2.C_h, 1)

poly1d_fn1 = np.poly1d(coeff1)
poly1d_fn2 = np.poly1d(coeff2)


plt.plot(lg_L14l, C_l, color = 'darkblue', marker = ".", markersize = 1, linestyle='None', alpha = 0.5)
plt.plot(lg_L14h, C_h, color = 'red', marker = ".", markersize = 2, linestyle='None')


plt.plot(ret1.lg_L14l,ret1.C_l, color = 'darkblue', marker = "o", markersize = 15, linestyle = 'None', alpha = 0.7)
plt.plot(ret2.lg_L14h,ret2.C_h, color = 'red', marker = "o", markersize = 15, linestyle = 'None', alpha = 0.7)


plt.errorbar(ret1.lg_L14l,ret1.C_l, color = 'blue', marker = "None", linestyle = 'None', yerr=poly1d_fn1(ret1.lg_L14l) - ret1.C_l, linewidth = 3, label = 'Low Excitation')
plt.errorbar(ret2.lg_L14h,ret2.C_h, color = 'red', marker = "None", linestyle = 'None', yerr=poly1d_fn2(ret2.lg_L14h) - ret2.C_h, linewidth = 3, label = 'High Excitation')


plt.xlabel('$log_{10}{(L_{1.4}}$ / $W Hz^{-1})$')
plt.ylabel('$C = (R_{90}$ / $R_{50}$)')
plt.grid()
plt.legend()
plt.xlim(17, 25)
plt.ylim(1.5, 4.5)
plt.show()
```
